```python
# Credit: this block of code courtesy of the starter code writer.

# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate unique counts of "school_name" and "student_name".
schools_unq_count = school_data_complete["school_name"].nunique()
student_unq_count = school_data_complete["Student ID"].nunique()





#Calculate total budget, referencing only school_data.
budget_tot = school_data["budget"].sum()

#Calculate average math and reading scores. Weighted by students.
math_av = school_data_complete["math_score"].mean()
read_av = school_data_complete["reading_score"].mean()

#Create a data frame of only those students passing math
only_math_pass = school_data_complete.loc[school_data_complete["math_score"] >= 70]

#Calculate the number of students passing math
math_pass_count = len(only_math_pass)


#Calculate the percentage of students passing math
percent_math_pass = math_pass_count / student_unq_count * 100

#Create a data frame of only those students passing reading
only_reading_pass = school_data_complete.loc[school_data_complete["reading_score"] >= 70]

#Calculate the number of students passing reading
reading_pass_count = len(only_reading_pass)

#Calculate the percentage of students passing reading
percent_reading_pass = reading_pass_count / student_unq_count * 100

#Calculate "Overall Passing Rate", defined in the problem as average of percent_reading_pass and 
# percent_math_pass
overall_pass = (percent_math_pass + percent_reading_pass) / 2


#Creating the district summary dictionary, the creating a data frame from that.
district_dict = {"Total Schools":[schools_unq_count], "Total Students":[student_unq_count],
                 "Total Budget":[budget_tot],
                 "Average Math Score": [math_av], "Average Reading Score":[read_av],
                "% Passing Math":[percent_math_pass], "% Passing Reading":[percent_reading_pass],
                "% Overall Passing Rate":[overall_pass]}

district_summary_df = pd.DataFrame(district_dict)

#Formatting Total Students with commas and Total Budget with commas and dollar signs

district_summary_df["Total Students"] = district_summary_df["Total Students"].astype(float).map("{:,.0f}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map("${:,.0f}".format)


district_summary_df

```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math

* Create a dataframe to hold the above results

```python
#school_data is a dataframe with a lot of this already in it, so I will
#copy it to school_summ and build on it

school_summ = school_data

school_summ["per_budg"] = school_summ["budget"] / school_summ["size"]

###



#Using groupby to group reading and math scores by school_name
school_scores = school_data_complete[["school_name","math_score", "reading_score"]].groupby(["school_name"])


#Using the groupby objects mean() method to create a dataframe of average math and reading scores
school_avs_df = school_scores.mean()

#Changing column names to what the numbers actually represent
school_avs_df = school_avs_df.rename(columns={"math_score":"Average Math Score", "reading_score":"Average Reading Score"})

school_avs_df
#scrap = school_data_complete.loc[school_data_complete["school_name"] == "Ford High School", :]


```

```python
#Creating a groubby object with only lines corresponding to students passing math
school_math_pass = school_data_complete[["school_name", "math_score"]].loc[school_data_complete["math_score"] >= 70 ]
school_math_pass_gb = school_math_pass.groupby("school_name")

#Creating dataframe with counts of math passers for each school
school_math_pass_count_df = school_math_pass_gb.count()


#Renaming the column to "Number Passing Math"
school_math_pass_count_df = school_math_pass_count_df.rename(columns ={"math_score":"Number Passing Math"})
school_math_pass_count_df

```

```python
#Creating a groubby object with only lines corresponding to students passing reading
school_reading_pass = school_data_complete[["school_name", "reading_score"]].loc[school_data_complete["reading_score"] >= 70 ]
school_reading_pass_gb = school_reading_pass.groupby("school_name")

#Creating dataframe with counts of reading passers for each school
school_reading_pass_count_df = school_reading_pass_gb.count()


#Renaming the column to "Number Passing Reading"
school_reading_pass_count_df = school_reading_pass_count_df.rename(columns ={"reading_score":"Number Passing Reading"})
school_reading_pass_count_df

```

```python
#At this point, there are four dataframes to merge

#Merge the numbers passing math and numbers passing reading
passing_counts_df = pd.merge(school_math_pass_count_df, school_reading_pass_count_df, on="school_name")
#passing_counts_df

pass_av_merge_df = pd.merge(passing_counts_df, school_avs_df, on="school_name")
#pass_av_merge_df

#Now merge this with school_summ, the summary so far
school_summ = pd.merge(school_summ, pass_av_merge_df, on="school_name")
school_summ
```

```python
#Translate numbers passing to percent passing, making new columns

school_summ["% Passing Math"] = school_summ["Number Passing Math"] / school_summ["size"] * 100
school_summ["% Passing Reading"] = school_summ["Number Passing Reading"] / school_summ["size"] * 100

#Conjure up lame "Overall Passing Rate"
school_summ["% Overall Passing Rate"] = (school_summ["% Passing Math"] + school_summ["% Passing Reading"]) / 2


school_summ
```

```python
#school_summ has a few too many columns, and some columns need renaming. That is done here.

display_summ = school_summ[["school_name", "type", "size", "budget", "per_budg", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
display_summ = display_summ.set_index("school_name")
display_summ.index.name = None


display_summ = display_summ.rename(columns={"type":"School Type", "size":"Total Students", "budget": "Total School Budget", "per_budg":"Per Student Budget"})


#display_summ["Total Students"] = display_summ["Total Students"].astype(float).map("{:,.0f}".format)
display_summ["Total School Budget"] = display_summ["Total School Budget"].astype(float).map("${:,.2f}".format)
display_summ["Per Student Budget"] = display_summ["Per Student Budget"].astype(float).map("${:,.2f}".format)

display_summ
```
